```python
from plm_entropy import embed_entropy_esm2, \
                        embed_entropy_protT5, \
                        aln_plm_entropy, \
                        calc_aln_entropy, \
                        al_mod_entr_correl

import pandas as pd
```

```python
#view input inference set:
h1alldf = pd.read_csv('IAV_HA_pLMs/testing_data/ncbi_h1_110424_filt/ncbi_h1_110424_filt_nodedetails.csv')
h1alldf
```

```python
#get a random sample of 500 external node sequences for demonstration
h1testdf = h1alldf[~h1alldf.node.str.contains('NODE_')].sample(n=500, random_state=42)
h1testdf.to_csv('ncbi_h1_110424_filt_nodedetails_sample.csv', index=False)
h1testdf
```

```python
#write fasta file of the sample sequences

with open('ncbi_h1_110424_filt_sample.fas', 'w') as out:
    out.write( '\n'.join([ '>%s\n%s'%(n,s) for n,s in zip(list(h1testdf.node),list(h1testdf.seq)) ]) )

#align to ncbi_h1_110424_filt_sample.aln.fas
```

```python
#embed with ESM-2 finetuned model
    #options:
        #path to the model
        #input file (csv or fasta)
        #output name

embed_entropy_esm2(
        '/media/spyros/HD-ADU3/spyros/flu_LLM_evol_data/models/esm2_t33-H1_110424_clu99_e10',
          'ncbi_h1_110424_filt_nodedetails_sample.csv',
           'ncbi_h1_110424_filt-esm2_H1'
                  )
```

```python
#embed with protT5 finetuned model
    #options:
        #path to the model
        #input file (csv or fasta)
        #output name

embed_entropy_protT5(
        '/media/spyros/HD-ADU3/spyros/model_weights_topublish/protT5MLM-H1/',
          'ncbi_h1_110424_filt_nodedetails_sample.csv',
           'ncbi_h1_110424_filt-protT5_H1'
                  )
```

```python
#match ESM2 model pLM entropy results to sequence alignment 
aln_plm_entropy(
    'ncbi_h1_110424_filt-esm2_H1-site_entropy.csv',
    'ncbi_h1_110424_filt_sample.aln.fas',
    'ncbi_h1_110424_filt-esm2_H1-site_entropy'
)
```

```python
#match protT5 model pLM entropy results to sequence alignment 
aln_plm_entropy(
    'ncbi_h1_110424_filt-protT5_H1-site_entropy.csv',
    'ncbi_h1_110424_filt_sample.aln.fas',
    'ncbi_h1_110424_filt-protT5_H1-site_entropy'
)
```

```python
#calculate the sequence alignment entropy
h1test_alnentr = calc_aln_entropy('ncbi_h1_110424_filt_sample.aln.fas')
```

```python
#calculate the spearman correlation between alignment entropy and ESM2 model pLM entropy
al_mod_entr_correl(
    h1test_alnentr,
    'ncbi_h1_110424_filt-esm2_H1-site_entropy.aln.csv',
)
```

```python
#calculate the spearman correlation between alignment entropy and protT5 model pLM entropy
al_mod_entr_correl(
    h1test_alnentr,
    'ncbi_h1_110424_filt-protT5_H1-site_entropy.aln.csv',
)
```
