```python
correl_list = []

for sero in ['ncbi_h1_110424_filt', 'ncbi_h3n2_110424_filt', 'gisaid_h5_270624_filt', 'gisaid_h7_270624_filt']:

    for mset in [
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2/esm2_reset/', 'base'],
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2_H1/checkpoint-1150/', 'H1'],
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2_H3/checkpoint-2130/', 'H3'],
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2_H5/checkpoint-441/', 'H5'],
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2_H7/checkpoint-160/', 'H7'],
                 ['/media/spyros/HD-ADU3/spyros/model_weights_topublish/reset_weights_ESM2/esm2_t33-weight_reset-HA_all_110424_clu99_e10/', 'HAall']
                ]:

        alnentr = calc_aln_entropy('/media/spyros/HD-ADU3/spyros/flu_LLM_evol_data/%s/ancestral_sequences.fasta'%sero, exclude_internal_nodes=True)
        
        r = al_mod_entr_correl(alnentr, '%s-esm2_reset_%s-site_entropy.aln.csv'%(sero,mset[1]), exclude_internal_nodes=True).statistic

        correl_list.append(['ESM-2-reset %s'%mset[1], '%s_R'%sero.split('_')[1].upper(), r ])

correl_df = pd.DataFrame(correl_list)
correl_df.columns = ['model', 'variable', 'value']
correl_df.to_csv('reset_weights_ESM2_correl.csv', index=False)
correl_df
```

## plotting

```python
correl_df = pd.read_csv('reset_weights_ESM2_correl.csv')
correl_df['model'] = [x.replace('-reset', '').replace(' base', '').replace('HAall', 'HA-all') for x in list(correl_df.model)]
correl_df['variable'] = [x.replace('H3N2', 'H3') for x in list(correl_df.variable)]
correl_df.columns = ['model', 'variable', 'reset']
correl_df
```

```python
orig_correl = pd.read_csv('../../../../plotting/esm2_correl_311024.tsv', sep='\t', index_col=0)
orig_correl = orig_correl[['H1_R', 'H3_R', 'H5_R', 'H7_R']]
orig_correl = orig_correl[:-2]
orig_correl.reset_index(drop=False, inplace=True)
orig_correl = orig_correl.melt(id_vars='index')
orig_correl.columns = ['model', 'variable', 'pretrained']
# orig_correl['modeltype'] = 'pretrained'
orig_correl

allcorrel = correl_df.merge(orig_correl)
allcorrel['correl_diff'] = allcorrel.pretrained - allcorrel.reset
allcorrel
```

```python
palette6 = ['#F257AC', '#47FF6C', '#04ADBF', '#F83530', '#BC7942', '#9794F2', '#6863F2']
```

```python
figmentr = px.bar(correl_df, y='value', x='variable', color='model',barmode='group', color_discrete_sequence =palette6, text='value')
figmentr.update_layout({'plot_bgcolor': 'rgba(0, 0, 0, 0)', 'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
figmentr.update_traces(texttemplate='%{text:.2f}',textposition='outside')
figmentr.update_yaxes(range=[-0.3, 1.1])
figmentr.update_layout( font=dict(size=20)) # You can adjust the size as needed
figmentr.update_layout(showlegend=False)
figmentr.write_html('reset_weights_ESM2_correl_140725_plot.html')
figmentr.write_image('reset_weights_ESM2_correl_140725_plot.svg', width=2000, height=500)
figmentr.show()
```

```python
figmentr = px.bar(allcorrel, y='correl_diff', x='variable', color='model',barmode='group', color_discrete_sequence =palette6, text='correl_diff')
figmentr.update_layout({'plot_bgcolor': 'rgba(0, 0, 0, 0)', 'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
figmentr.update_traces(texttemplate='%{text:.2f}',textposition='outside')
figmentr.update_yaxes(range=[-1.1, 1.1])
figmentr.update_layout( font=dict(size=20)) # You can adjust the size as needed
figmentr.update_layout(showlegend=False)
figmentr.write_html('reset_weights_ESM2_correl_140725_diff_plot.html')
figmentr.write_image('reset_weights_ESM2_correl_140725_diff_plot.svg', width=2000, height=500)
figmentr.show()
```
